```python
from google.cloud import storage
import pandas as pd
import io


def load_data_from_storage(storage_client, TARGET_BUCKET_NAME, filename):
    '''
        Descarga archivo .csv desde bucket, lo carga y retorna como dataframe
    '''
    # Descargamos archivo
    try:
        bucket = storage_client.bucket(TARGET_BUCKET_NAME)
        blob = bucket.blob(filename)
        file_bytes = blob.download_as_bytes()

        print(f'Archivo {filename} descargado exitosamente.')
    except Exception as err:
        print('Error descargando archivo.')
        print(err)
        return None, False

    # Cargamos archivo como dataframe y lo retornamos
    try:
        file_like_object = io.StringIO(file_bytes.decode('utf-8'))
        partial_column_data_types = {'Conversation_id': str, 'ANI': str, 'Segmento': str, 'Rut_empresa': str, "Tipo_de_cliente": str, "EPA": str, "Resolucion": str, "Transferidas_a_ejecutivo": str}
        dataframe = pd.read_csv(file_like_object, dtype=partial_column_data_types)
        return dataframe, True

    except Exception as err:
        print('Error retornando como dataframe archivo.')
        print(err)
        return None, False
```
